2. Most Selling Product.
   - By Quantity
   - By Revenue

```python
data.head()
```

```python
data.groupby('Product')['Quantity'].sum()
```

```python
data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
most_quantity = data.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
most_quantity
```

```python
type(most_quantity)
```

```python
most_quantity = most_quantity.reset_index()
most_quantity
```

```python
type(most_quantity)
```

```python
plt.figure(figsize=(9,4))
plt.bar(most_quantity['Product'], most_quantity['Quantity'], color=['red','blue','yellow','green', 'pink']);
plt.title("Most Selling Product - By Quantity")
plt.xlabel("Product")
plt.ylabel("Quantity");
```

```python
data['Revenue'] = data['Price'] * data['Quantity']
```

```python
data.head()
```

```python
data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
```

```python
most_revenue = data.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
```

```python
most_revenue = most_revenue.reset_index()
```

```python
most_revenue
```

```python
plt.figure(figsize=(10,4))
plt.bar(most_revenue.Product, most_revenue.Revenue, color=['red','blue','yellow','green', 'pink']);
plt.title("Most Selling Product - By Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue");
```
